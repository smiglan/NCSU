```
from os import listdir
from numpy import asarray
from numpy import vstack
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
from numpy import savez_compressed
import pandas as pd
path = '/content/gdrive/My Drive/542 projectmydata/unltestl/'

def load_images(path, size=(128,128)):
  src_list = list()
  i=0
  for filename in listdir(path):
    # load and resize the image
    pixels = load_img(path + filename, target_size=size)
    # convert to numpy array
    pixels = img_to_array(pixels)
    print(filename)
    src_list.append(pixels)
    i=i+1

  return [asarray(src_list)]

# dataset path
X = load_images(path)
```

```
import numpy as np
from sklearn.model_selection import train_test_split
X = np.array(X)
X = np.reshape(X,(4418,128,128,3))
print(X.shape)
print('Loaded: ', X.shape)
# save as compressed numpy array
filename = 'unlabelleddata.npz'
savez_compressed('/content/gdrive/My Drive/542ProjectC/'+filename, X)
print('Saved dataset: ', filename)
print()
```
